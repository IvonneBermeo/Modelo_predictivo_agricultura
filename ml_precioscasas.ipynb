### 4. Modelo predictivo con regresi√≥n log√≠stica

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix


# Agrupar las categor√≠as de LOCALITY
def group_locality(locality):
    if locality in ['New York', 'New York County', 'Queens County', 'Kings County', 
                    'Bronx County', 'Richmond County', 'United States']:
        return locality
    else:
        return 'Otros'

df['LOCALITY_GROUPED'] = df['LOCALITY'].apply(group_locality)

# Paso 1: Preprocesamiento de los datos
# Eliminar filas con valores faltantes
df = df.dropna()

# Convertir variables categ√≥ricas en variables dummy
categorical_columns = ['BROKERTITLE', 'TYPE', 'ADDRESS', 'STATE', 'MAIN_ADDRESS', 
                       'ADMINISTRATIVE_AREA_LEVEL_2', 'SUBLOCALITY', 'STREET_NAME', 
                       'LONG_NAME', 'FORMATTED_ADDRESS', 'LOCALITY_GROUPED']
df = pd.get_dummies(df, columns=categorical_columns, drop_first=True)

# Paso 2: Crear una variable objetivo binaria
# Definimos el umbral de "caro" como el percentil 75 del precio
threshold = df['PRICE'].quantile(0.75)
df['PRICE_CATEGORY'] = (df['PRICE'] >= threshold).astype(int)

# Paso 3: Separar las caracter√≠sticas (X) y la variable objetivo (y)
X = df[['PROPERTYSQFT', 'LATITUDE', 'LONGITUDE'] + list(df.columns[df.columns.str.startswith('BROKERTITLE_')])]
y = df['PRICE_CATEGORY']

# Escalar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Paso 4: Entrenar el modelo de regresi√≥n log√≠stica con mayor n√∫mero de iteraciones
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Paso 5: Evaluar el modelo
y_pred = model.predict(X_test)

# Reporte de clasificaci√≥n
print(classification_report(y_test, y_pred))

# Matriz de confusi√≥n
print(confusion_matrix(y_test, y_pred))
```

### Reporte de Clasificaci√≥n:

Precisi√≥n (Precision): La precisi√≥n para la clase 0 es del 88% y para la clase 1 es del 67%. Esto significa que el 88% de las casas clasificadas como "no caras" y el 67% de las casas clasificadas como "caras" son realmente clasificadas correctamente.

Recall: El recall para la clase 0 es del 92%, lo que indica que el 92% de las casas "no caras" fueron identificadas correctamente. Para la clase 1, el recall es del 57%, lo que significa que solo el 57% de las casas "caras" fueron identificadas correctamente.

F1-score: El F1-score, que es una medida ponderada de la precisi√≥n y el recall, es del 90% para la clase 0 y del 62% para la clase 1. Esto indica que el modelo tiene un mejor rendimiento en la predicci√≥n de casas "no caras" que en casas "caras".

Support: El soporte es el n√∫mero de observaciones en cada clase. Hay 740 observaciones de casas "no caras" (clase 0) y 221 observaciones de casas "caras" (clase 1).


### Interpretaci√≥n General:


La precisi√≥n general del modelo es del 84%, lo que indica que el 84% de las predicciones realizadas por el modelo son correctas.
Estos resultados indican que el modelo puede ser √∫til para predecir si una casa no es cara, pero puede tener dificultades para identificar correctamente las casas caras.
